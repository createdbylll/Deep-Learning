```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('train.csv')
```

```python
data
```

```python
df = data.copy()
```

```python
df['IDF_TYP_CD'] = df['IDF_TYP_CD'].apply(lambda row : int(row[2:]))
```

```python
df.drop('CUST_ID', inplace=True, axis=1)
```

```python
def getvalue(x):
    if(x == 'N'):
        return 0
    elif(x == 'Y'):
        return 1
    else:
        return x
for i in df.columns:
    df[i] = df[i].apply(getvalue)
```

```python
def getgender(x):
    if(x == 'X'):
        return 1
    else:
        return int(x)
df['GENDER'] = df['GENDER'].apply(getgender)
```

```python
# 归一化
for i in df.columns:
    print(i)
    df[i] = df[i].astype(float)
    df[i] = (df[i]-df[i].min())/(df[i].max()-df[i].min())
```

```python
delete_list = []
for i in df.columns:
    if(pd.isna(df[i][0])):
        delete_list.append(i)
```

```python
for i in delete_list:
    df.drop(i, inplace=True, axis=1)
```

```python
# 将DataFrame写入CSV文件，使用制表符作为分隔符
df.to_csv('processed_data.csv', sep='\t', index=False)
```
